```python
# import des librairies
from pathlib import Path
import zipfile

import pandas as pd
import soccerdata as sd
```

```python
# création des dossiers de données

RAW_DIR = Path("../data/raw")

KAGGLE_DIR = RAW_DIR / "kaggle"
SOCCER_DIR = RAW_DIR / "soccerdata"

KAGGLE_DIR.mkdir(parents=True, exist_ok=True)
SOCCER_DIR.mkdir(parents=True, exist_ok=True)

print("Dossiers créés.")
```

```python
# téléchargement du dataset FIFA World Cup depuis Kaggle

import os

dataset = "piterfm/fifa-football-world-cup"

os.system(
    f"kaggle datasets download -d {dataset} -p {KAGGLE_DIR}"
)
```

```python
# extraction du fichier zip

zip_file = list(KAGGLE_DIR.glob("*.zip"))[0]

with zipfile.ZipFile(zip_file, "r") as zip_ref:
    zip_ref.extractall(KAGGLE_DIR)

print("Dataset Kaggle extrait.")
```

```python
# voir les fichiers extraits

list(KAGGLE_DIR.glob("*"))
```

```python
# connexion a la base de données SoccerData

fb = sd.FBref(
    leagues="INT-World Cup",
    seasons=[2014, 2018, 2022]
)
```

```python
# Calendrier des matchs 

schedule = fb.read_schedule()

schedule.to_csv(
    SOCCER_DIR / "schedule.csv",
    index=False
)

schedule.head()
```

```python
# statistiques des joueurs
players = fb.read_player_season_stats(
    stat_type="standard"
)

players.to_csv(
    SOCCER_DIR / "player_stats.csv",
    index=True
)

players.head()
```

```python
# statistiques offensives
shooting = fb.read_player_season_stats(
    stat_type="shooting"
)

shooting.to_csv(
    SOCCER_DIR / "player_shooting.csv"
)
```

```python
# statistiques des passes

passing = fb.read_player_season_stats(
    stat_type="passing"
)

passing.to_csv(
    SOCCER_DIR / "player_passing.csv"
)
```

```python
# statistiques des défenses
defense = fb.read_player_season_stats(
    stat_type="defense"
)

defense.to_csv(
    SOCCER_DIR / "player_defense.csv"
)
```

```python
# statistiques des possessions

possession = fb.read_player_season_stats(
    stat_type="possession"
)

possession.to_csv(
    SOCCER_DIR / "player_possession.csv"
)
```

```python
# statistiques des équipes

team_stats = fb.read_team_season_stats(
    stat_type="standard"
)

team_stats.to_csv(
    SOCCER_DIR / "team_stats.csv"
)
```
